```python
# ===============================
# Patch-based Lightweight U-Net training for semantic segmentation
# Dataset: semantic-segmentation-of-aerial-imagery
# Simpler/faster architecture: Reduced filters (32-64-128-256-512)
# ===============================

import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras import layers, models
import random

# Check for GPU (enable in Kaggle: Settings > Accelerator > GPU)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using {len(gpus)} GPU(s)")
    except RuntimeError as e:
        print(e)
else:
    print("No GPU found. Training on CPU (slower).")

# -------------------------------
# Dataset settings
# -------------------------------
root_directory = "/kaggle/input/semantic-segmentation-of-aerial-imagery"
patch_size = 256  # Reduce to 128 for 4x faster (fewer pixels)
max_images = 20  # Subsample for speed; set to None for full dataset (~180 images)

# Classes (hex -> RGB; order: Building=0, Land=1, Road=2, Vegetation=3, Water=4, Unlabeled=5)
class_colors = {
    "Building": "#3C1098",
    "Land": "#8429F6",
    "Road": "#6EC1E4",
    "Vegetation": "#FEDD3A",
    "Water": "#E2A929",
    "Unlabeled": "#9B9B9B"
}

# Convert HEX to RGB array
for key in class_colors:
    hex_val = class_colors[key].lstrip('#')
    class_colors[key] = np.array([int(hex_val[i:i+2], 16) for i in (0, 2, 4)])

# -------------------------------
# Helper function: create patches
# -------------------------------
def create_patches(img, patch_size):
    """
    Split an image (H,W,C) into non-overlapping patches of size (patch_size, patch_size).
    """
    h, w = img.shape[:2]
    ph, pw = patch_size, patch_size
    # Crop to make divisible by patch_size
    img = img[:h - (h % ph), :w - (w % pw)]
    patches = []
    for i in range(0, img.shape[0], ph):
        for j in range(0, img.shape[1], pw):
            patch = img[i:i + ph, j:j + pw]
            patches.append(patch)
    return patches

# -------------------------------
# Collect paired image/mask paths
# -------------------------------
image_paths = []
mask_paths = []

for path, _, files in os.walk(root_directory):
    dirname = os.path.basename(path)
    if dirname == 'images':
        for f in files:
            if f.lower().endswith(('.tif', '.tiff', '.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(path, f))
    elif dirname == 'masks':
        for f in files:
            if f.lower().endswith(('.tif', '.tiff', '.png', '.jpg')):
                mask_paths.append(os.path.join(path, f))

# Sort paths for pairing (assumes filenames correspond when sorted; adjust if needed, e.g., by stem)
image_paths.sort()
mask_paths.sort()

# Truncate to max_images and ensure equal length
num_pairs = min(len(image_paths), len(mask_paths), max_images or float('inf'))
image_paths = image_paths[:num_pairs]
mask_paths = mask_paths[:num_pairs]

if num_pairs == 0:
    raise ValueError("No paired images/masks found. Check dataset structure and file extensions.")

print(f"Using {num_pairs} paired images/masks.")

# -------------------------------
# Load and process paired patches
# -------------------------------
image_patches_all = []
mask_patches_all = []
scaler = MinMaxScaler()

for idx, (img_path, mask_path) in enumerate(zip(image_paths, mask_paths)):
    try:
        # Load image
        image = cv2.imread(img_path)
        if image is None:
            print(f"Warning: Failed to load {img_path}")
            continue
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32)

        # Load mask
        mask = cv2.imread(mask_path)
        if mask is None:
            print(f"Warning: Failed to load {mask_path}")
            continue
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)

        # Crop to matching size
        h, w = min(image.shape[0], mask.shape[0]), min(image.shape[1], mask.shape[1])
        image = image[:h, :w]
        mask = mask[:h, :w]

        # Crop to multiple of patch_size
        size_y = (image.shape[0] // patch_size) * patch_size
        size_x = (image.shape[1] // patch_size) * patch_size
        image = image[:size_y, :size_x]
        mask = mask[:size_y, :size_x]

        if size_y == 0 or size_x == 0:
            print(f"Warning: Image {img_path} too small for patches.")
            continue

        # Create patches
        img_patches = create_patches(image, patch_size)
        mask_patches = create_patches(mask, patch_size)

        # Collect image patches for global scaling
        image_patches_all.extend(img_patches)
        mask_patches_all.extend(mask_patches)

        if (idx + 1) % 5 == 0:
            print(f"Processed {idx + 1}/{num_pairs} images ({len(img_patches)} patches each)")

    except Exception as e:
        print(f"Error processing pair {idx}: {e}")
        continue

# Convert to NumPy arrays
image_dataset = np.array(image_patches_all)
mask_dataset = np.array(mask_patches_all)

print(f"Total patches created: {len(image_dataset)}")

if len(image_dataset) == 0:
    raise ValueError("No patches created. Check image sizes or paths.")

# Global normalization: Fit scaler once on all image pixels
all_pixels = image_dataset.reshape(-1, 3)
scaler.fit(all_pixels)
image_dataset = scaler.transform(all_pixels).reshape(image_dataset.shape)

# -------------------------------
# Convert RGB masks to integer labels (0-5)
# -------------------------------
def rgb_to_label(mask_rgb):
    label = np.zeros(mask_rgb.shape[:2], dtype=np.uint8)
    for idx, color in enumerate(class_colors.values()):
        match = np.all(mask_rgb == color, axis=-1)
        label[match] = idx
    # Fallback: Set unmatched pixels to Unlabeled (class 5)
    label[~np.isin(label, list(range(len(class_colors))))] = 5
    return label

labels = np.array([rgb_to_label(m) for m in mask_dataset])
labels = np.expand_dims(labels, -1)

# One-hot encode labels
n_classes = len(class_colors)
labels_cat = to_categorical(labels, num_classes=n_classes)

# -------------------------------
# Train/test split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    image_dataset, labels_cat, test_size=0.2, random_state=42, stratify=None  # No stratify for multi-dim y
)
print(f"Train patches: {len(X_train)}, Test patches: {len(X_test)}")

# -------------------------------
# Build Lightweight U-Net model (simpler/faster: fewer filters)
# -------------------------------
def lightweight_unet(input_size=(patch_size, patch_size, 3), n_classes=6):
    inputs = layers.Input(input_size)

    # Encoder (reduced filters: 32-64-128-256)
    c1 = layers.Conv2D(32, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(32, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(64, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(64, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(128, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(128, 3, activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = layers.Conv2D(256, 3, activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(256, 3, activation='relu', padding='same')(c4)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bottleneck (reduced: 512 vs 1024)
    c5 = layers.Conv2D(512, 3, activation='relu', padding='same')(p4)
    c5 = layers.Conv2D(512, 3, activation='relu', padding='same')(c5)

    # Decoder (symmetric, with skip connections)
    u6 = layers.Conv2DTranspose(256, 2, strides=2, padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(256, 3, activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(256, 3, activation='relu', padding='same')(c6)

    u7 = layers.Conv2DTranspose(128, 2, strides=2, padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(128, 3, activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(128, 3, activation='relu', padding='same')(c7)

    u8 = layers.Conv2DTranspose(64, 2, strides=2, padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(64, 3, activation='relu', padding='same')(u8)
    c8 = layers.Conv2D(64, 3, activation='relu', padding='same')(c8)

    u9 = layers.Conv2DTranspose(32, 2, strides=2, padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(32, 3, activation='relu', padding='same')(u9)
    c9 = layers.Conv2D(32, 3, activation='relu', padding='same')(c9)

    # Output: Softmax for multi-class segmentation
    outputs = layers.Conv2D(n_classes, 1, activation='softmax')(c9)

    model = models.Model(inputs=[inputs], outputs=[outputs])
    return model

# Compile model
model = lightweight_unet(input_size=(patch_size, patch_size, 3), n_classes=n_classes)
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()  # ~7.5M parameters

# -------------------------------
# Train the model
# -------------------------------
history = model.fit(
    X_train, y_train,
    batch_size=32,  # Increased for speed (reduce if OOM)
    epochs=50,  # Increase to 10-20 for better performance
    verbose=1,
    validation_data=(X_test, y_test)
)

# -------------------------------
# Plot training curves
# -------------------------------
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------------
# Predict on a random test patch
# -------------------------------
if len(X_test) > 0:
    idx = random.randint(0, len(X_test) - 1)
    test_img = X_test[idx]
    gt_mask = np.argmax(y_test[idx], axis=-1)

    pred = model.predict(np.expand_dims(test_img, 0), verbose=0)
    pred_mask = np.argmax(pred, axis=-1)[0]

    # Visualize
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.title("Input Patch")
    plt.imshow(test_img)
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Ground Truth Mask")
    plt.imshow(gt_mask, cmap='tab10', vmin=0, vmax=n_classes-1)
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Predicted Mask")
    plt.imshow(pred_mask, cmap='tab10', vmin=0, vmax=n_classes-1)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("No test data available for prediction.")
```

```python
def predict_full_image(image_path, model, patch_size=256):
    # Load & preprocess
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32)

    # Crop to multiple of patch_size
    h, w = image.shape[:2]
    h_crop = (h // patch_size) * patch_size
    w_crop = (w // patch_size) * patch_size
    image = image[:h_crop, :w_crop]

    # Normalize using same scaler as training
    flat = image.reshape(-1, 3)
    flat = scaler.transform(flat)
    image = flat.reshape(image.shape)

    # Split into patches
    patches = create_patches(image, patch_size)
    patches = np.array(patches)

    # Predict each patch
    preds = model.predict(patches, verbose=0)
    preds = np.argmax(preds, axis=-1)

    # Reconstruct full mask
    mask_pred = np.zeros((h_crop, w_crop), dtype=np.uint8)
    idx = 0
    for i in range(0, h_crop, patch_size):
        for j in range(0, w_crop, patch_size):
            mask_pred[i:i+patch_size, j:j+patch_size] = preds[idx]
            idx += 1

    return image, mask_pred
```

```python
test_image_path = image_paths[0]  # or any external image path
img, pred_mask = predict_full_image(test_image_path, model, patch_size)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Input Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Predicted Segmentation")
plt.imshow(pred_mask, cmap="tab10", vmin=0, vmax=n_classes-1)
plt.axis("off")
plt.show()
```
